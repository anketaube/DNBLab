```python
import requests
import pandas as pd
from datetime import datetime

# Erstelle eine leere Liste, um alle Testergebnisse zu speichern
testergebnisse = []

def test_url(url, area):
    now = datetime.now()
    status_code = None
    status = None

    try:
        response = requests.get(url)
        status_code = response.status_code
        status = "Erfolgreich" if response.ok else f"Fehler: {status_code}"

        # Speichere das Testergebnis in die Liste
        testergebnisse.append({'Bereich': area, 'URL': url, 'Zeitpunkt des Aufrufs': now, 'Statuscode': status_code, 'Status': status})

    except requests.exceptions.RequestException as e:
        status = f"Fehler: {e}"
        # Speichere das Testergebnis in die Liste
        testergebnisse.append({'Bereich': area, 'URL': url, 'Zeitpunkt des Aufrufs': now, 'Statuscode': status_code, 'Status': status})

# Liste der Bereiche und URLs
bereiche_urls = [
    ("Applikationen", "https://portal.dnb.de/bookviewer/view/1035757001#page/n0/mode/2up"),
    ("Applikationen", "https://portal.dnb.de/audioplayer/do/show/107677184X"),
    ("IHVZ", "https://d-nb.info/1276542003/04"),
    ("NP", "https://d-nb.info/1284464601/34"),
    ("Applikationen", "https://portal.dnb.de/emulation/view/1165906597"),
    ("Webarchive", "https://webarchiv.dnb.de/show.aspx?GUID=3ababb18-78d0-eb11-94bd-d4ae528b7600&PARENT=0ff4c3c7-da3a-eb11-8831-d4ae528b7600"),
    ("Webarchive", "https://web.archive.org/web/20050214021402/http://winter.einrichten-preisvergleich.de"),
    ("OAI Schnittstelle", "http://services.dnb.de/oai2/repository?verb=GetRecord&identifier=oai:dnb.de/dnb:digitalisate-oa:projekt3/1049398327&metadataPrefix=mets"),
    ("OAI Schnittstelle", "https://services.dnb.de/oai/repository?verb=Identify"),
    ("SRU Schnittstelle", "https://services.dnb.de/sru/dnb?operation=explain&version=1.1"),
    ("SRU Schnittstelle", "https://services.dnb.de/sru/cbs?version=1.1&operation=searchRetrieve&query=PER%3DSchneider*"),
    ("SRU Schnittstelle", "https://services.dnb.de/sru/dnb?version=1.1&operation=searchRetrieve&query=PER%3DSchneider*"),
    ("SRU Schnittstelle", "https://services.dnb.de/sru/zdb?version=1.1&operation=searchRetrieve&query=PER%3DSchneider*"),
    ("SRU Schnittstelle", "https://services.dnb.de/sru/authorities?version=1.1&operation=searchRetrieve&query=PER%3DSchneider*"),
    ("URN Resolver", "https://nbn-resolving.org/"),
    ("URN Resolver", "https://nbn-resolving.org/institutions"),
    ("URN Resolver", "https://nbn-resolving.org/urn:nbn:de:gbv:547-200700768"),
    ("URN Resolver", "https://nbn-resolving.org/urn:nbn:de:bvb:12-bsb00113387-4")
]

# Durchlaufe die Liste und teste jede URL
for bereich, url in bereiche_urls:
    test_url(url, bereich)

# Erstelle den DataFrame mit allen Testergebnissen
df = pd.DataFrame(testergebnisse)

# Apply style to the DataFrame for better readability
styled_df = df.style.set_table_styles([{'selector': 'tr:hover','props': [('background-color', 'yellow')]}])

# Zeige den DataFrame als formatierte Tabelle in Jupyter Notebook an
display(styled_df)
```
