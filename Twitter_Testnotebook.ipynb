```python
def get_all_texts(row_element): 

    tweettexts = []
    
    for item in row_element: 
        first_layer = pd.DataFrame(item)
        for text in first_layer.text:
            tweettexts.append(text)

    return tweettexts
        
    
```

```python
texts = get_all_texts(df.data)
```
