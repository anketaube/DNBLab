# BEDO-Test Datentöpfe

Es werden nacheinander digitale Objekte von den verschiedenen Anwendungen aus verschiedenen Datentöpfen aufgerufen. 

Externer Aufruf:

```python
import requests
import pandas as pd
from datetime import datetime

# Erstelle eine leere Liste, um alle Testergebnisse zu speichern
testergebnisse = []

def test_url(area, repo, url):
    now = datetime.now()
    status_code = None
    status = None

    try:
        response = requests.get(url)
        status_code = response.status_code
        status = "Erfolgreich" if response.ok else f"Fehler: {status_code}"

        # Speichere das Testergebnis in die Liste
        testergebnisse.append({'Bereich': area, 'Datentopf' : repo,  'URL': url, 'Zeitpunkt des Aufrufs': now, 'Statuscode': status_code, 'Status': status})

    except requests.exceptions.RequestException as e:
        status = f"Fehler: {e}"
        # Speichere das Testergebnis in die Liste
        testergebnisse.append({'Bereich': area, 'Datentopf' : repo, 'URL': url, 'Zeitpunkt des Aufrufs': now, 'Statuscode': status_code, 'Status': status})

# Liste der Bereiche und URLs
bereiche_urls = [
    ("Bookviewer -DBM;", "bookscan", "https://portal.dnb.de/bookviewer/view/1280840676"),
    ("Audioplayer - HTT", "dma","https://portal.dnb.de/audioplayer/do/show/107677184X"),
    ("TOC - PDF","toc2", "https://d-nb.info/1276542003/04"),
    ("TOC - TXT", "toc2", "https://d-nb.info/1276542003/04/text"),
    ("TOC - TIF", "toc2", "https://d-nb.info/1276542003/04/tiff"),
    ("TOC - PDF", "Yuuvis", "http://etc.dnb.de/repo-as-bes/api/objekte/[IDN]"),
    ("NP - ZIP/EPUB","rome", "https://d-nb.info/1284464601/34"),
    ("NP - ZIP/EPUB","netpub2", "https://d-nb.info/1284464601/34"),
    ("NP - PDF","netpub2", " https://d-nb.info/996026665/34"),
    ("EMiL", "multimedia", "https://portal.dnb.de/emulation/view/1165906597"),
    ("Webarchiv - Website", "bookscan", "https://webarchiv.dnb.de/show.aspx?GUID=3ababb18-78d0-eb11-94bd-d4ae528b7600&PARENT=0ff4c3c7-da3a-eb11-8831-d4ae528b7600"),
    ("Webarchiv - Zeitschnitt", "bookscan",  "https://web.archive.org/web/20050214021402/http://winter.einrichten-preisvergleich.de"),
    ("OAI2 - Get Record", "bookscan", "http://services.dnb.de/oai2/repository?verb=GetRecord&identifier=oai:dnb.de/dnb:digitalisate-oa:projekt3/1049398327&metadataPrefix=mets"),
    ("EPAS - PDF", "rome", "http://etc.dnb.de/epas/api/issues/idn/1006393633/content")    
]

# Durchlaufe die Liste und teste jede URL
for bereich, repo, url in bereiche_urls:
    test_url(bereich, repo, url)

# Erstelle den DataFrame mit allen Testergebnissen
df = pd.DataFrame(testergebnisse)
left_aligned_df = df.style.set_properties(**{'text-align': 'left'}) 

# Text in Spalten links ausrichten

left_aligned_df = left_aligned_df.set_table_styles( 
[dict(selector = 'th', props=[('text-align', 'left')])]) 

# Zeige den DataFrame als formatierte Tabelle in Jupyter Notebook an
display(left_aligned_df) 
```
