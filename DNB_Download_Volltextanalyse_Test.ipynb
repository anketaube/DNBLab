Eingabe eines Suchwortes: 

```python
from IPython.display import display 
import ipywidgets as widgets 
from ipywidgets import interact, Layout 

tb1 = widgets.Text(value = 'Amerika', description='Stichwort: ');display(tb1);

# ein Button zur Übernahme der eingegebenen Werte 
button = widgets.Button(description='Übernehmen!', layout=Layout(width='200px')); 
button.style.button_color = 'lightgreen';display(button); 
def on_button_clicked(sender): 
    a = int(tb1.value)
    print('Suchhwort: ' + str(tb1.value))  
    button.on_click(on_button_clicked)
```

Einfügen der IDN_OCR.txt und ausführen der Suche mit übernommenen Suchwort:

```python
filename = '1026552451_OCR.txt'
search = str(tb1.value)
with open(filename) as f:
    for num, line in enumerate(f, 1):
        if search in line:
            print('%s - found at line:' % search, num)
```
